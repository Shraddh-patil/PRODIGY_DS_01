Dropping unnecessary Columns

```python
df.drop(['Indicator Name'],axis=1,inplace=True)
```

Separating regions from

```python
# Merge population data with metadata to get the 'Region' column
# Metadata contains 'Country Code' and 'Region'
# First, rename the columns in the metadata DataFrame to match the expected names
metadata_renamed = metadata.rename(columns={'AFE': 'Country Code', 'Unnamed: 1': 'Region'})
df_merged = pd.merge(df, metadata_renamed[['Country Code', 'Region']], on='Country Code')

# FILTER: Keep only rows where 'Region' is NOT empty
# Real countries have a region; aggregates (World, Europe, etc.) do not
df_countries = df_merged[df_merged['Region'].notna()]
```

Top 10 countries w.r.t Population

```python
#Get Top 10 Countries for 2024
top_ten = df_countries.sort_values(by="2024", ascending=False).head(10)

print("Top 10 Actual Countries:")
print(top_ten[['Country Name', 'Country Code', '2024']])
```

Bar plot for total population in the year 2024 and the year 2020

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2024", y="Country Code", data=top_ten, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2024)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y="Country Code", data=top_ten, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2020)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

Extracting bottom 10 countries w.r.t. population

```python
bottom_ten = df_countries.sort_values(by="2024", ascending=True).head(10)

print("Bottom 10 Actual Countries:")
print(bottom_ten[['Country Name','Country Code','2024']])
```

Bottom 10 countries for the year 2024 and 2020

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2024", y="Country Code", data=bottom_ten, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2024)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y="Country Code", data=bottom_ten, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2020)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
